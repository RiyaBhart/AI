```python
import random

class Enviroment:
    def __init__(self):
        # self.system = ['Vunerable','Safe','Safe','Vunerable','Vunerable','Safe','Vunerable','Safe','Safe']
        self.system = [random.choice(['safe', 'vulnerable']) for _ in range(9)]


    def get_percept(self, position):
        return self.system[position]

    def patch(self,position):
        self.system[position]='Safe'

    def display_system(self,agent_position):
        print("Current state of system\n")
        system_with_agent = self.system[:]
        system_with_agent[agent_position]="*"
        for i in range(10):
            print(" | ".join(system_with_agent[i:i+1]))
```

```python
class SimpleReflexAgent:
    def __init__(self):
        self.position = 0

    def act(self,percept,system):
        if percept == 'vulnerable':
            system[self.position]='Safe'
            return "position is vulnerable"
        else:
            return "position is safe"

    def move(self):
        if self.position<=9:
            self.position+=1
        return self.position
```

```python
def RunAgent(agent,enviroment,steps):
    for step in range(steps):
        percept = enviroment.get_percept(agent.position)
        action = agent.act(percept,enviroment.system)
        print(f"\nStep {step + 1}: Position {agent.position} -> Percept- {percept}, Action - {action}")
        enviroment.display_system(agent.position)
        if percept == 'vulnerable':
            enviroment.patch(agent.position)
        agent.move()

agent = SimpleReflexAgent()
enviroment = Enviroment()

RunAgent(agent,enviroment,9)
```
